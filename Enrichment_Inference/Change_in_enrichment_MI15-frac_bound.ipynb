```python
L = 16
mutation_names = ['N26D','S29R','Y35N','Y48C','D49Y','V98I','G31D','Y33H','M34I','H35N','N52H','G57D','L83V','S84N','R85G','R87K']

df = pd.read_csv('20220723_fracbound_v2.csv',
                 dtype={"geno":"str"})
df["variant_int"] = df.geno.apply(lambda x: int(x, 2))

df = df.set_index("variant_int").reindex(range(0, 2**L)).reset_index()
sequences_int = range(0,2**L)
print(sequences_int)
for ii in range(L):
    df[f"mutation_{ii+1}"] = df.geno.str[ii] == '1'


dft = df.set_index("geno")
dfpl = pd.DataFrame()
exp = dft.MI_frac_mean.to_dict()
errexp = dft.MI_frac_sem.to_dict()
nbmuts, deltas, muts, keeps, deltas_err, sources, targets = [], [], [], [], [], [], []
for im, m in tqdm(enumerate(mutation_names)):
    muts += [m]*(len(df))
    for s in df.geno:
        deltas += [exp[s[:im] + '1' + s[im+1:]] - exp[s]]
        deltas_err += [np.sqrt(errexp[s]**2 + errexp[s[:im] + '1' + s[im+1:]]**2)]
        nbmuts += [s.count('1')]
        keeps += [s[im] != '1']
        sources += [s]
        targets += [s[:im] + '1' + s[im+1:]]
dfpl = pd.DataFrame()
dfpl["Mutation"] = muts
dfpl[r'dfrac'] = deltas
dfpl[r'Number of mutations'] = nbmuts
dfpl[r'dfrac_sem'] = deltas_err
dfpl[r'source'] = sources
dfpl[r'target'] = targets
dfpl = dfpl[keeps]
```
